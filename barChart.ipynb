```python
import matplotlib.pyplot as plt
import numpy as np

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
labels = ['CUB', 'CAR', 'SOP', 'In-shop', 'Hotel']
orgi = [34.7, 21.8, 81.0, 89.9, 4.5]
semi = [56.5, 67.9, 81.1, 90.6, 18.8]
hneg = [57.7, 73.3, 81.9, 90.9, 29.2]


x = np.arange(len(labels))  # the label locations
width = 0.275  # the width of the bars
plt.figure(figsize=(8,8))
rects0 = plt.bar(x - width, orgi, width, label='HN')
rects1 = plt.bar(x, semi, width, label='SHN')
rects2 = plt.bar(x + width, hneg, width, label='SCT')
autolabel(rects0)
autolabel(rects1)
autolabel(rects2)
plt.xticks(x,labels,fontsize=20)
plt.yticks(fontsize=15)
plt.ylabel('R@1',fontsize=20)
plt.title('Recall@1 Comparison',fontsize=24)
plt.legend(fontsize=15)
plt.savefig('bar',dpi=200)
plt.show()
```
